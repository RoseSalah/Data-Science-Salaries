```python
import pandas as pd
import os
```

Reading data set from csv

```python
data = pd.read_csv("ds_salaries.csv")
```

Discovering data

```python
data.shape
data.columns
data.head()
```

Discovering Null and Duplicates: 

```python
data.isnull().sum()
data.duplicated().sum() #1171 duplicated out of the  total 3755 rows
```

```python
# data[data.duplicated()]
data = data.drop_duplicates()
data.shape

```

```python
data.drop(data[['salary','salary_currency']], axis=1, inplace=True)
# we can use the 'salay' and 'salary_currency' columns to find the salary in USD, but we don't need them for our analysis,
# 'salary_in_usd' column is enough
print(data.shape)
```

As discovered from the data above, The data has 11 columns and more than 3700 rows. 
I dropped salary and salary currency columns as they are not needed where the salay_in_usd column is enough. 
The data has no null values! 
So we ended up with 9 columns of the data, each providing specific information related to the topic.


```python
# value of each columns: 
# for column in data.columns:
#     print(f"{column}: {data[column].nunique()}")    
    
# replacing experience level with full form
data['experience_level'] = data['experience_level'].replace({
    'EN': 'Entry Level', 
    'MI': 'Mid Level', 
    'SE': 'Senior Level', 
    'EX': 'Executive Level'
})
# replacing employment type with full form
data['employment_type'] = data['employment_type'].replace({
    'PT': 'Part Time', 
    'FT': 'Full Time', 
    'CT': 'Contract', 
    'FL': 'Freelance'
})
# replacing company size with full form
data['company_size'] = data['company_size'].replace({
    'S': 'Small', 
    'M': 'Medium', 
    'L': 'Large'
})
# data.head(30)
```

```python
data.remote_ratio = data.remote_ratio.replace({
    0: 0.0, 
    50: 0.5, 
    100: 1
})
data.head(30)
```

```python
# from object to string
data['experience_level'] = data['experience_level'].astype('string') 
data['employment_type'] = data['employment_type'].astype('string')
data['job_title'] = data['job_title'].astype('string')
data['employee_residence'] = data['employee_residence'].astype('string')
data['company_location'] = data['company_location'].astype('string')
data['company_size'] = data['company_size'].astype('string')
```
